```python
import pandas as pd
```

```python
twoHand = ['A','B','D', 'F', 'G', 'H', 'K', 'M', 'N', 'P', 'Q', 'S', 'T', 'W', 'X', 'Y']
```

```python
df[df['LABEL'].isin(twoHand)].groupby(['LABEL']).count()
```

```python
df[~df['LABEL'].isin(twoHand)].groupby(['LABEL']).count()
```
